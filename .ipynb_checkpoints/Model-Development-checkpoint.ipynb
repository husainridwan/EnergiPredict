##### Getting the total enegy consumption feature

```python
def calc_weights(df, sw_load, nw_load):
  # Check for potential division by zero
  if df[sw_load].sum() + df[nw_load].sum() == 0:
    raise ValueError("Total HVAC load cannot be zero")

  # Calculate total load for each row
  df['total_load'] = df[sw_load] + df[nw_load]
  df['hvac_total'] = (df[sw_load] / df['total_load']) * df[sw_load] + (df[nw_load] / df['total_load']) * df[nw_load]

  # Drop temporary column (optional)
  df.drop(['total_load', 'hvac_S', 'hvac_N'], axis=1, inplace=True)
  return df
```

```python
df = calc_weights(data, 'hvac_S', 'hvac_N')
df.head()
```

```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.model_selection import KFold, cross_val_score, train_test_split, RandomizedSearchCV
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
def split(X, y, test_size=0.2, val_size=0.2):
  if test_size + val_size > 1:
    raise ValueError("The sum of test_size and val_size must be less than or equal to 1.")

  # Split data into training and testing sets
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

  if val_size > 0:
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=val_size, random_state=42)
  else:
    X_val, y_val = None, None

  return X_train, X_val, X_test, y_train, y_val, y_test
```

```python
X_train, X_val, X_test, y_train, y_val, y_test = split(X, y, test_size=0.2, val_size=0.1)
```

```python
def tune_model(model, param_dist, X_train, y_train, X_val, y_val, scoring="f1", n_iter=10, cv=5):
    random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, scoring=scoring, n_iter=n_iter, cv=cv)
    random_search.fit(X_train, y_train)
    
    # Get the best model and best parameters
    best_model = random_search.best_estimator_
    best_params = random_search.best_params_
    
    if X_val is not None and y_val is not None:
        cv_scores = cross_val_score(best_model, X_val, y_val, scoring=scoring, cv=cv)
    else:
        cv_scores = None
    
    return best_model, best_params, cv_scores
```

```python
# Define the parameter distributions for each model
param_dist_rf = {'n_estimators': [100, 200], 'max_depth': [10, 20, 30]}
param_dist_xgb = {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1, 1.0]}
param_dist_lgbm = {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1, 1.0]}
param_dist_cat = {'iterations': [100, 200], 'learning_rate': [0.01, 0.1, 1.0]}

# Tune the models
best_model_rf, best_params_rf, cv_scores_rf = tune_model(RandomForestRegressor(), param_dist_rf, X_train, y_train, X_val, y_val, scoring="neg_mean_squared_error")
best_model_xgb, best_params_xgb, cv_scores_xgb = tune_model(XGBRegressor(), param_dist_xgb, X_train, y_train, X_val, y_val, scoring="neg_mean_squared_error")
best_model_lgbm, best_params_lgbm, cv_scores_lgbm = tune_model(LGBMRegressor(), param_dist_lgbm, X_train, y_train, X_val, y_val, scoring="neg_mean_squared_error")
best_model_cat, best_params_cat, cv_scores_cat = tune_model(CatBoostRegressor(verbose=0), param_dist_cat, X_train, y_train, X_val, y_val, scoring="neg_mean_squared_error")

# Make predictions on the test data using the best models
y_pred_rf = best_model_rf.predict(X_test)
y_pred_xgb = best_model_xgb.predict(X_test)
y_pred_lgbm = best_model_lgbm.predict(X_test)
y_pred_cat = best_model_cat.predict(X_test)

# Print the best parameters and cross-validation scores for each model
print("Best parameters for Random Forest Regressor: ", best_params_rf)
print("Cross-validation scores for Random Forest Regressor: ", cv_scores_rf)
print("Best parameters for XGBoost: ", best_params_xgb)
print("Cross-validation scores for XGBoost: ", cv_scores_xgb)
print("Best parameters for LightGBM: ", best_params_lgbm)
print("Cross-validation scores for LightGBM: ", cv_scores_lgbm)
print("Best parameters for CatBoost: ", best_params_cat)
print("Cross-validation scores for CatBoost: ", cv_scores_cat)
```

```python
from sklearn.ensemble import BaggingRegressor, GradientBoostingRegressor, StackingRegressor, RandomForestRegressor

# Define the parameter distributions for each model
param_dist_bag = {'n_estimators': [100, 200]}
param_dist_boost = {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1]}
param_dist_stack = {'final_estimator__alpha': [0.01, 0.1]}

# Tune the Bagging model
best_model_bag, best_params_bag, cv_scores_bag = tune_model(BaggingRegressor(), param_dist_bag, X_train, y_train, X_val, y_val, scoring="neg_mean_squared_error")

# Tune the Boosting model
best_model_boost, best_params_boost, cv_scores_boost = tune_model(GradientBoostingRegressor(), param_dist_boost, X_train, y_train, X_val, y_val, scoring="neg_mean_squared_error")

# Define base models for Stacking
base_models = [('xgb', XGBRegressor(n_estimators=200, learning_rate=0.1)), 
               ('lgbm', LGBMRegressor(n_estimators=200, learning_rate=0.1)), 
               ('cat', CatBoostRegressor(learning_rate=0.1, iterations=200))]
# base_models = [('xgb', XGBRegressor(**best_params_xgb)), 
#                ('lgbm', LGBMRegressor(**best_params_lgbm)), 
#                ('rf', RandomForestRegressor())


# Tune the Stacking model
best_model_stack, best_params_stack, cv_scores_stack = tune_model(StackingRegressor(estimators=base_models, final_estimator=Ridge()), 
                                                                  param_dist_stack, X_train, y_train, X_val, y_val, scoring="neg_mean_squared_error")

# Make predictions on the test data using the best models 
y_pred_bag = best_model_bag.predict(X_test)
y_pred_boost = best_model_boost.predict(X_test)
y_pred_stack = best_model_stack.predict(X_test)

# Print the best parameters and cross-validation scores for each model
print("Best parameters for Bagging: ", best_params_bag)
print("Cross-validation scores for Bagging: ", cv_scores_bag)
print("Best parameters for Boosting: ", best_params_boost)
print("Cross-validation scores for Boosting: ", cv_scores_boost)
print("Best parameters for Stacking: ", best_params_stack)
print("Cross-validation scores for Stacking: ", cv_scores_stack)
```

```python
import pickle
# Train and evaluate each model using the best parameters
models = {'XGBoost': XGBRegressor(**best_params_xgb),
          'LightGBM': LGBMRegressor(**best_params_lgbm),
          'RandomForest': RandomForestRegressor(**best_params_rf),
          'CatBoost': CatBoostRegressor(**best_params_cat),
          'Bagging': BaggingRegressor(**best_params_bag),
          'Boosting': GradientBoostingRegressor(**best_params_boost)}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred_val = model.predict(X_val)
    print(f"{name} validation MSE: {mean_squared_error(y_val, y_pred_val)}")
    pickle.dump(model, open(f'{name}_model.pkl', 'wb'))
 
# Define base models for Stacking
base_models = [('xgb', XGBRegressor(**best_params_xgb)), 
               ('lgbm', LGBMRegressor(**best_params_lgbm)), 
               ('cat', CatBoostRegressor(**best_params_cat))]

# Train and evaluate the Stacking model
stack_model = StackingRegressor(estimators=base_models, final_estimator=Ridge(alpha=best_params_stack['final_estimator__alpha']))
stack_model.fit(X_train, y_train)
y_pred_val = stack_model.predict(X_val)
print(f"Stacking validation MSE: {mean_squared_error(y_val, y_pred_val)}")
pickle.dump(stack_model, open('stack.pkl', 'wb'))

# Make predictions on the test data using the trained models
for name, model in models.items():
    y_pred_test = model.predict(X_test)
    print(f"{name} test predictions: {y_pred_test}")

# Make predictions on the test data using the Stacking model
y_pred_test = stack_model.predict(X_test)
print(f"Stacking test predictions: {y_pred_test}")
```
